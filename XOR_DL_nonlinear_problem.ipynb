```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
model = Sequential()
model.add(Dense(2, input_dim=2, activation='relu'))  # hidden layer with 2 neurons
model.add(Dense(1, activation='sigmoid'))            # output layer with 1 neuron
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=100, verbose=0)
```

```python
_, accuracy = model.evaluate(X, y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
predictions = model.predict(X)
predictions = np.round(predictions).astype(int)
```

```python
print("Predictions:")
for i in range(len(X)):
    print(f"Input: {X[i]} => Predicted Output: {predictions[i]}, Actual Output: {y[i]}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# XOR input and output
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])  # just 1D array, simpler

# Build the model
model = Sequential([
    Dense(2, input_shape=(2,), activation='relu'),  # hidden layer with 2 neurons
    Dense(1, activation='sigmoid')                   # output layer
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model (shorter epochs but should still work mostly)
model.fit(X, y, epochs=2000, verbose=0)

# Predict and print results
predictions = model.predict(X)
predictions = (predictions > 0.5).astype(int)  # convert probs to 0 or 1

for i in range(len(X)):
    print(f"Input: {X[i]} => Predicted: {predictions[i][0]}, Actual: {y[i]}")
```
